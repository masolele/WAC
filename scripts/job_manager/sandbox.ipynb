```python
import geopandas as gpd
import pandas as pd

gpkg = "WAC_CNA_CommodityAoI_GridsCollationUpdate.gpkg"
out = "WAC_CNA_CommodityAoI_GridsCollationUpdate_all_layers.parquet"

layers = [
    "WAC_BelCNAaoiUTM_CoffeeBRA_Tiles",
    "WAC_BelCNAaoiUTM_CoffeeHND_Tiles",
    "WAC_BelCNAaoiUTM_TimberCMR_Tiles",
    "WAC_CNAaoiUTM_RubberTHA_Tiles",
    "WAC_GerCNAaoiUTM_Cattle_Tiles",
    "WAC_GerCNAaoiUTM_Cocoa_Tiles",
    "WAC_GerCNAaoiUTM_Coffee_Tiles",
    "WAC_GerCNAaoiUTM_OilPalm_Tiles",
    "WAC_GerCNAaoiUTM_Rubber_Tiles",
    "WAC_GerCNAaoiUTM_Soy_Tiles",
    "WAC_GerCNAaoiUTM_Timber_Tiles",
    "WAC_ItaCNAaoiUTM_CoffeeETH_Tile",
    "WAC_NldCNAaoiUTM_CocoaGHA_Tile",
    "WAC_NldCNAaoiUTM_CocoaLBR_Tiles",
    "WAC_NldCNAaoiUTM_TimberPNG_Tile",
    "WAC_NldCNAaoiUTM_TimberPNG_Tiles1",
    "WAC_RouCNAaoiUTM_TimberROU_Tiles",
]

gdfs = []

for layer in layers:
    print(f"Reading {layer}...")

    gdf = gpd.read_file(gpkg, layer=layer)
    gdf["source_layer"] = layer

    gdfs.append(gdf)

combined = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True), crs=gdfs[0].crs)

combined.to_parquet(out, index=False)

print(f"\nSaved {len(combined):,} features to:")
print(out)
```

```python
import geopandas as gpd

gdf = gpd.read_parquet(
    "WAC_CNA_CommodityAoI_GridsCollationUpdate_all_layers.parquet"
).reset_index(drop=True)

gdf = gdf[gdf.geometry.notna() & ~gdf.geometry.is_empty].copy()
gdf["geometry"] = gdf.geometry.make_valid()
gdf = gdf.explode(index_parts=False).reset_index(drop=True)
gdf = gdf[gdf.geometry.geom_type.isin(["Polygon", "MultiPolygon"])].copy()
gdf = gdf[gdf.geometry.notna() & ~gdf.geometry.is_empty].copy()

tol = 1  # CRS units, probably meters

tmp = gdf[["geometry"]].copy()
tmp["geometry"] = tmp.geometry.buffer(0).buffer(tol)

pairs = gpd.sjoin(tmp, tmp, how="inner", predicate="intersects")
pairs = pairs[pairs.index != pairs["index_right"]]

parent = list(range(len(gdf)))


def find(x):
    while parent[x] != x:
        parent[x] = parent[parent[x]]
        x = parent[x]
    return x


def union(a, b):
    ra, rb = find(a), find(b)
    if ra != rb:
        parent[rb] = ra


for left, right in zip(pairs.index, pairs["index_right"]):
    union(left, right)

gdf["adjacency_group"] = [find(i) for i in range(len(gdf))]

filtered = (
    gdf.sort_index()
    .drop_duplicates("adjacency_group")
    .drop(columns="adjacency_group")
    .reset_index(drop=True)
)

filtered.to_parquet("tiles_one_per_adjacent_group.parquet", index=False)
```
