```python
import geopandas as gpd 

gdf = gpd.read_file("AOI_ESA_WAC_HND.gpkg")

```

```python
import pandas as pd

gdf = gdf.to_crs("EPSG:3857")

gdf['spatial_extent'] = gdf.geometry.apply(lambda x: {'west': x.bounds[0], 'south': x.bounds[1], 'east': x.bounds[2], 'north': x.bounds[3], 'crs': 'EPSG:3857'})

gdf['temporal_extent'] = [['2020-01-01', '2020-12-31']] * len(gdf)

gdf['id'] = gdf.index.astype(str)

gdf['year'] = '2020'

gdf['crs'] = 'EPSG:3857'

gdf['model_id'] = 'WorldAgriCommodities_LatinAmerica_v1'

gdf = gdf[['geometry', 'id', 'spatial_extent', 'temporal_extent', 'crs', 'model_id', 'year']]

# Create a copy
gdf_2025 = gdf.copy()

# Update temporal extent
gdf_2025['temporal_extent'] = [['2025-01-01', '2025-12-31']] * len(gdf_2025)

# (Optional but recommended) update IDs to keep them unique
gdf_2025['id'] = (gdf.index + len(gdf)).astype(str)

gdf_2025['year'] = '2025'

# Combine
gdf = pd.concat([gdf, gdf_2025], ignore_index=True)

gdf 

```

```python
gdf.to_parquet('job_database_backup.parquet', index=False)
```
